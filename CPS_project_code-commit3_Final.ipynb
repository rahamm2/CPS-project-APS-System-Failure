```python
# Checking the missing in train data and sort them from higher to lower
def missing(df):
    print (round((df.isnull().sum() * 100/ len(df)),2).sort_values(ascending=False))
missing(data_train)
```

```python
# Dropping the missing value columns  
def remove_nan(df, threshold):
    percentage_null = df.isnull().sum() / len(df)
    missing_features = percentage_null[percentage_null > threshold].index
    data=df.drop(missing_features, axis=1, inplace=True)
    return data 
```

```python
# Remove nan value over 35% threshold from train data 
remove_nan(data_train,.35)
```

### Identify Bin Histogram

```python
#Identify the bin and Count the Bin
from collections import Counter
Temp=[]

for item in X_train.columns:
    Temp.append(item.split('_')[0])
    counter=Counter(Temp)
feature_prefix=list(counter.keys())
print(feature_prefix)
number_bin=(list(counter.values()))
print(number_bin)
```

```python
## Identify Hitogram bin identifier
hist_identify=[]
for x,y in enumerate(number_bin):
    if y==10:
        hist_identify.append(feature_prefix[x])
print(hist_identify)
```

```python
# Identify Histogram feature
Histogram_features=[z for z in X_train.columns if z.split('_')[0] in hist_identify]
print(Histogram_features)
print("The total number of feature", len(Histogram_features))
```

### Feature Selection

```python
# Importing necessary Libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
```

```python
numer_vars = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numerical = list(X_train.select_dtypes(include=numer_vars).columns)
X_train =X_train[numerical]
X_train.shape
```

```python
numer_vars= ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numerical = list(X_test.select_dtypes(include=numer_vars).columns)
X_test =X_test[numerical]
X_test.shape
```

```python
# removing corelated features
def corr(data, thres):
    col_corr = set()  # Set correlated columns
    corr_matrix = data.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > thres: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features= corr(X_train, 0.8)
print('correlated features for the dataset: ', len(set(corr_features)) )
```

```python
# removing corellated features features
X_train.drop(labels=corr_features, axis=1, inplace=True)
X_test.drop(labels=corr_features, axis=1, inplace=True)

X_train.shape, X_test.shape
```

```python
# Using step forward feature top 10 features for my model

from mlxtend.feature_selection import SequentialFeatureSelector as SFS

sfs = SFS(RandomForestRegressor(), 
           k_features=10, 
           forward=True, 
           floating=False, 
           verbose=2,
           scoring='r2',
           cv=3)

sfs = sfs.fit(np.array(X_train), y_train)
```
