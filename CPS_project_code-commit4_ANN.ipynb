### Missing Value Treatment for Training and Testing Data

```python
#checking total missing values
data_train.isnull().sum().sum()
```

```python
# Checking missing values by collumn wises
missing_value_count=data_train.isnull().sum()
missing_value_count
```

```python
# Checking the missing in train data and sort them from higher to lower
def missing(df):
    print (round((df.isnull().sum() * 100/ len(df)),2).sort_values(ascending=False))
missing(data_train)
```

```python
# Dropping the missing value columns  
def remove_nan(df, threshold):
    percentage_null = df.isnull().sum() / len(df)
    missing_features = percentage_null[percentage_null > threshold].index
    data=df.drop(missing_features, axis=1, inplace=True)
    return data 
```

```python
# Remove nan value over 35% threshold from train data 
remove_nan(data_train,.35)
```

### Identify Bin Histogram

```python
#Identify the bin and Count the Bin
from collections import Counter
Temp=[]

for item in X_train.columns:
    Temp.append(item.split('_')[0])
    counter=Counter(Temp)
feature_prefix=list(counter.keys())
print(feature_prefix)
number_bin=(list(counter.values()))
print(number_bin)
```

```python
## Identify Hitogram bin identifier
hist_identify=[]
for x,y in enumerate(number_bin):
    if y==10:
        hist_identify.append(feature_prefix[x])
print(hist_identify)
```

```python
# Identify Histogram feature
Histogram_features=[z for z in X_train.columns if z.split('_')[0] in hist_identify]
print(Histogram_features)
print("The total number of feature", len(Histogram_features))
```

### Feature Selection

```python
##Import necessary libraries for best feature selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
# instantiate SelectKBest to determine 30 best features
best_features = SelectKBest(score_func=f_classif, k=30)
fit = best_features.fit(X_train,y_train)
df_scores = pd.DataFrame(fit.scores_)
df_columns = pd.DataFrame(X_train.columns)
```

```python
# concatenate dataframes
feature_scores = pd.concat([df_columns, df_scores],axis=1)
feature_scores.columns = ['Feature_Name','Score']  # name output columns
```

```python
print(feature_scores.nlargest(30,'Score'))  # print 30 best features
```

```python
# Creating dataframe with select feature
cols = best_features.get_support(indices=True)
X_train_selected_feature=X_train.iloc[:,cols]
```

```python
X_train_selected_feature
```

```python
X_test_top_feature=X_test.iloc[:,cols]
```

```python
X_test_top_feature 
```

### Balancing the dataset

```python
## Performing oversampling and importing the file
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
```

```python
## Performing oversampling
rs=RandomOverSampler(.75)
X_train_RS, y_train_RS=rs.fit_resample(X_train_selected_feature,y_train)
print("class before fit{}".format(Counter(y_train)))
print("class after fit{}".format(Counter(y_train_RS)))
```

## Building ANN model without Hyperparameter Tuning

```python
## Inport Tensorflow library
import tensorflow as tf
```

```python
#Initializing the ANN
ann=tf.keras.models.Sequential()
## Adding first hidden layer  and input layer 
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
## Adding second hidden layer
ann.add(tf.keras.layers.Dense(units=12,activation='relu'))
## Adding third hidden layer
ann.add(tf.keras.layers.Dense(units=12,activation='relu'))
##Aing output layer
ann.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
```

```python
#training ANN 
## Compiling the ANN
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
##Training the Ann
ann.fit(X_train_RS,y_train_RS,batch_size=32,epochs=100)
```

### Predicting test result

```python
y_pred=ann.predict(X_test_top_feature)
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
confusion_matrix=confusion_matrix(y_test,y_pred)
print("Confusion matrix: \t")
print(confusion_matrix)
confusion_matrix= pd.DataFrame(confusion_matrix.reshape((1,4)), columns=['TN', 'FP', 'FN', 'TP'])
tc = 10*confusion_matrix.FP + 500*confusion_matrix.FN
print(f'Total cost accrue: {float(tc.values[0])}')
print("The accuracy score:",accuracy_score(y_test,y_pred)*100)
print(classification_report(y_test,y_pred))
```

## **************Building ANN model For Hyperparameter Tuning*********************


```python
def build_ann(unit):
  # creating the layers of the ANN
  ann = tf.keras.models.Sequential()
  ann.add(tf.keras.layers.Dense(units=unit, activation='relu'))
  ann.add(tf.keras.layers.Dense(units=unit, activation='relu'))
  ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
  ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
  return ann
```

```python
# Creating the object for Keras classifier
# Import Keras classifier library
from keras.wrappers.scikit_learn import KerasClassifier
model=KerasClassifier(build_fn=build_ann)
```

### Hypermeter tuning for ANN

```python
from sklearn.model_selection import GridSearchCV
params={'batch_size':[100, 20, 50, 25, 32], 
        'nb_epoch':[200, 100, 300, 400],
        'unit':[5,6, 10, 11, 12, 15],
           
        }
gs=GridSearchCV(estimator=model, param_grid=params, cv=10)
# now fit the dataset to the GridSearchCV object. 
gs = gs.fit(X_train_RS,y_train_RS)
```

```python
print(" The best parameter:",gs.best_params_)
```

```python
print(" The best accuracy:",gs.best_score_)
```

### Predicting the Test set results

```python
y_pred=gs.predict(X_test_top_feature)
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
confusion_matrix=confusion_matrix(y_test,y_pred)
print("Confusion matrix: \t")
print(confusion_matrix)
confusion_matrix= pd.DataFrame(confusion_matrix.reshape((1,4)), columns=['TN', 'FP', 'FN', 'TP'])
tc = 10*confusion_matrix.FP + 500*confusion_matrix.FN
print(f'Total cost accrue: {float(tc.values[0])}')
print("The accuracy score:",accuracy_score(y_test,y_pred)*100)
print(classification_report(y_test,y_pred))
```
